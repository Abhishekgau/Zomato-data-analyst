```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
df = pd.read_csv('Zomato data .csv',lineterminator ='\n')
```

### To remove the rate columns inside the '/' symbol 

```python
def handleRate(value):
    value = str(value).split('/')
    value= value[0];
    return float(value)
   
df['rate'] = df['rate'].apply(handleRate)
```

```python
print(df.head())
```

```python
df.info()
```

## Visualization part start

### what type of restaurant do the majority of customers order from

```python
df.head()
```

```python
figure = figsize=(8,5)
sns.countplot(x=df['listed_in(type)'],palette = 'Set2')
plt.xlabel('type')

plt.show()
```

### conclusion
majority of the restaurant falls in dinning  ctegory

### how many votes has each type of restaurant received from customers

```python
df.head()
```

```python
grouped_data = df.groupby('listed_in(type)')['votes'].sum()
result = pd.DataFrame({'votes':grouped_data})
plt.plot(result,c= 'green',marker = 'o')
plt.xlabel('Type of restaurant', c= 'red', size= 20)
plt.ylabel('Votes',c= 'red', size = 20)
```

### conclusion
dinning restaurant has received maximum votes

 ### what are the rating that the majority of restaurants have received

```python
plt.hist(df['rate'],bins = 20)
plt.title('Rating of the majority')
plt.show()
```

### conclusion
The majority  rastaurant received rating from 3.5 to 4

### Which mode(online or offline) has received the maximum rating

```python
plt.figure(figsize=(6,6))
sns.boxplot(x = 'online_order',y='rate', data = df)
plt.show()
```

### conclusion
offline oredr received lower rating in comparison to online order

### which type of restaurant received more offline orders so that zomato can  customer with some good offers

```python
pivot_table = df.pivot_table(index = 'listed_in(type)', columns = 'online_order', aggfunc = 'size', fill_value = 0)
sns.heatmap(pivot_table,annot = True, cmap= 'YlGnBu', fmt = 'd')
plt.title('Heatmap')
plt.xlabel('online Order')
plt.ylabel('Listed in(type)')
plt.show()
```

### conclusion
Dinnig restaurant primarily accept offline orderd whereas cafe primarily receive online order.This suggest that client preferes
order in person at restaurant but prefer online ordering at cafe
